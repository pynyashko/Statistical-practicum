### Problem 1

```python
from matplotlib.lines import lineStyles
import numpy as np 
import matplotlib.pyplot as plt

def f(x):
    return 2*np.sin(x)-np.cos(2*x)


def df(x):
    return 2*np.cos(x)+2*np.sin(2*x)


def tangent(x, x0):
    return f(x0) + df(x0)*(x-x0)


def beauty_derivative_plot(x0):
    if -2*np.pi < x0 < 2*np.pi:
        x = np.linspace(-2*np.pi, 2*np.pi, 1000)
        x_for_tangent = np.linspace(x0-1, x0+1, 1000)
        y = f(x)
        dy = df(x)
        tangent_line = tangent(x_for_tangent, x0)
        plt.figure(figsize=(14,10))
        plt.plot(x, y, "b-", label = "f(x) = 2sinx - cos2x")
        plt.plot(x, dy, "r--", label = "f(x) = 2cosx + 2sin2x")
        plt.plot(x_for_tangent, tangent_line, "g-", label = "Tangent at x = %0.2f"%x0)
        plt.plot(x0, f(x0), "ko")
        plt.vlines(x0, 0, f(x0), colors = "purple", linestyles = "dashed")
        plt.hlines(f(x0), 0, x0, colors = "purple", linestyles = "dashed")

        plt.title("Function, Derivative and Tangent Line")
        plt.annotate("f(%0.2f) = %0.2f"%(x0, f(x0)), xy = (x0, f(x0)), 
                                xytext = (x0-.5, f(x0)-1.5),
                                arrowprops={"facecolor": "black",
                                            "width": 0.2,
                                            "headwidth": 5,
                                            "connectionstyle": "arc3, rad=0.3"})
        
        
        plt.legend(loc = "upper right")
        plt.xlabel("x")
        plt.axhline(0, color="black", linewidth = 1)
        plt.ylabel("y")
        plt.axvline(0, color="black", linewidth = 1)
        plt.axis([-2*np.pi, 2*np.pi, -4, 4])
        plt.grid(True)
        plt.show
        return True
    print(f"X is out of range")
    return False


beauty_derivative_plot(-3)
```

### Problem 2

```python
dataset = np.loadtxt("../datasets/populations.txt", dtype = int)
years = dataset[:,0]
hares = dataset[:,1]
lynx = dataset[:,2]
carrots = dataset[:,3]
#print(np.concatenate((hares, lynx, carrots)))

plt.figure(figsize=(10, 4))
plt.grid(True, which="both")
plt.axis([min(years), max(years), 0, max(np.concatenate((hares, lynx, carrots))) + 5e3])
plt.xticks(years)

# Population graph
plt.plot(years, hares, "b-", label = "Hares")
plt.plot(years, lynx, "g-", label = "Lynx")
plt.plot(years, carrots, "r-", label = "Carrots")
plt.legend(loc = "best")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.grid(True, which="both")
plt.axis([min(years), max(years), 0, max(np.concatenate((hares, lynx, carrots))) + 5e3])
plt.xticks(years)

# Largest population
plt.scatter(years[np.argmax(hares)], hares[np.argmax(hares)], color = "blue")
plt.text(years[np.argmax(hares)], hares[np.argmax(hares)] + 3e3, "Largest hares population", color = "blue")
plt.scatter(years[np.argmax(lynx)], lynx[np.argmax(lynx)], color = "green")
plt.text(years[np.argmax(lynx)], lynx[np.argmax(lynx)] + 3e3, "Largest lynx population", color="green")
plt.scatter(years[np.argmax(carrots)], carrots[np.argmax(carrots)], color = "red")
plt.text(years[np.argmax(carrots)], carrots[np.argmax(carrots)] + 3e3, "Largest carrots population", color="red")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.grid(True, which="both")
#plt.xticks(years)
plt.xlabel("Year")
plt.ylabel("Population ratio, %")

total = hares + lynx
hares_ratio = (hares / total) * 100
lynx_ratio = (lynx / total) * 100 

plt.bar(years, hares_ratio, color = "blue", label = "Hares")
plt.bar(years, lynx_ratio, bottom = hares_ratio, color = "green", label = "Lynx")
plt.legend()


plt.show()
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(311)
plt.hist(lynx, bins = 10, rwidth=0.9, color = "green", label = "Lynx")
plt.axvline(np.mean(lynx), color = "red", linewidth = 3)
plt.text(np.mean(lynx) + 1000, 3.5, "Mean value")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.legend()

plt.subplot(312)
plt.hist(hares, bins = 10, rwidth=0.9, color = "blue", label = "Hares")
plt.axvline(np.mean(hares), color = "red", linewidth = 3)
plt.text(np.mean(hares) + 1000, 3, "Mean value")
plt.xlabel("Population")
plt.ylabel("frequency")
plt.legend()

plt.subplot(313)
plt.hist(carrots, bins = 10, rwidth=0.9, color = "red", label = "Carrots")
plt.axvline(np.mean(carrots), color = "black", linewidth = 3)
plt.text(np.mean(carrots) + 100, 3, "Mean value")
plt.xlabel("Population")
plt.ylabel("frequency")
plt.legend()

plt.show()
```

```python
lynx_win = 0
hares_win = 0
print("[", end = "")
i = 0
for i in range(len(years)-1):
    if lynx[i] > hares[i]:
        lynx_win += 1
        print("L", end = ", " )
    else:
        hares_win += 1
        print("H", end = ", ")
if lynx[i + 1] > hares[i + 1]:
    print("L", end = "")
else:
    print("H", end="")
print("]")

plt.pie([lynx_win, hares_win], labels = ["Lynx", "Hares"], colors=["green", "blue"])

```

```python
lynx_change = np.diff(lynx)
hares_change = np.diff(hares)
years_change = years[1:]

plt.figure(figsize=(10, 4))
plt.grid(True, which="both")
plt.axis([min(years_change), max(years), min(np.concatenate((hares_change, lynx_change)))-5e3, max(np.concatenate((hares_change, lynx_change))) + 5e3])
plt.xticks(years_change)
plt.axhline(0, color = "black", linewidth = 1)

# Population graph
plt.plot(years_change, hares_change, "b-", label = "Hares change")
plt.plot(years_change, lynx_change, "g-", label = "Lynx change")

plt.legend(loc = "best")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show

corr_coef = np.corrcoef(hares_change, lynx_change)[0,1]
print(f"Correlation coefficient: {corr_coef}")
```

```python
hares_train = hares[:-1]
lynx_train = lynx[:-1]

a, b = np.polyfit(lynx_train, hares_train, 1)

lynx_1920 = lynx[-1]
hare_est = a * lynx_1920 + b

hare_true = hares[-1]

error_ratio = abs(hare_est - hare_true)

print("Estimated hare population:", hare_est)
print("True hare population:", hare_true)
print("Error:", error_ratio)
```
